Cleaned the data and changed the datatypes for better results.

```python
df['Gross'] = df['Gross'].str.replace('$','')
df['Gross'] = df['Gross'].str.replace('M','')
df['Gross'] = df['Gross'].astype(float)
```

```python
df['VOTES'] = df['VOTES'].str.replace(',','')
df['VOTES'] = df['VOTES'].fillna(0)
df['VOTES'] = df['VOTES'].astype(int)
```

```python
#boxplot to see how the categorical feature “Species” is distributed with all other four input variables

fig, axes = plt.subplots(2, 2, figsize=(16,9))
sns.boxplot( y='RATING', data=df, orient='v' , ax=axes[0, 0], palette='YlGnBu', showfliers=False)
sns.boxplot( y='Gross',  data=df, orient='v' , ax=axes[0, 1], palette='YlGnBu', showfliers=False)
sns.boxplot( y='VOTES', data=df, orient='v' , ax=axes[1, 0], palette='YlGnBu', showfliers=False)
sns.boxplot( y='RunTime', data=df, orient='v' , ax=axes[1, 1], palette='YlGnBu', showfliers=False)
plt.show()
```

```python
#The violin plot shows density of the length and width in the species.
#The thinner part denotes that there is less density whereas the fatter part conveys higher density

fig, axes = plt.subplots(2, 2, figsize=(16,10))
sns.violinplot( y='RATING', data=df, orient='v' , ax=axes[0, 0], palette='YlGnBu',inner='quartile')
sns.violinplot( y='RunTime', data=df, orient='v' , ax=axes[0, 1], palette='YlGnBu',inner='quartile')
sns.violinplot( y='VOTES', data=df, orient='v' , ax=axes[1, 0], palette='YlGnBu',inner='quartile')
sns.violinplot( y='Gross', data=df, orient='v' , ax=axes[1, 1], palette='YlGnBu',inner='quartile')
plt.show()
```

The distribution for 'RATINGS' can be seen with a bulk in the range from 6 to 8 suggesting multiple ratings between 6 and 8.

The distribution 'RunTime' shows multiple values in the range of 0 to 200. Suggesting multiple movies are of the duration from 0 to 200 mins and very few movies are of more duration.This distribution has multiple peaks suggesting  the presence of multiple modes in the data.

For the 'VOTES' distribution , suggests multiple votes in the range of 0 and 0.25 particularly.

While the 'Gross' distribution shows the presence of outliers which are data points that fall significantly outside the main body of the distribution.

```python
df['GENRE'] = df['GENRE'].str.replace('\n', '', regex=True).str.strip()
```

```python
#Checking Mean & Median Values for each species
df.groupby('GENRE').agg(['mean', 'median'])
```

```python
sns.set(style="whitegrid")
top_genres = df['GENRE'].value_counts().head(5).index

# Create a FacetGrid for the distribution of 'RATING'
g = sns.FacetGrid(df[df['GENRE'].isin(top_genres)], hue="GENRE", height=5)
g.map(sns.kdeplot, "RATING", fill=True, warn_singular=False)
g.add_legend(title='GENRE', labels=top_genres)

plt.title(' RATING vs GENRE')
plt.show()
```

The above distribution shows the distribution for ratings for the top 5 Genre.Since these are the top 5 genres , the ratings for these are quite higher

```python
sns.set(style="whitegrid")

selected_genre_df = df[df['GENRE'].str.contains('Action') & ~df['GENRE'].isna()]
sns.distplot(selected_genre_df['RATING'].dropna(), hist=True, color='magenta')
plt.title(f'Distribution of RATING for Action')
plt.show()
```

```python
sns.set(style="whitegrid")

selected_genre_df = df[df['GENRE'].str.contains('Thriller') & ~df['GENRE'].isna()]
sns.distplot(selected_genre_df['RATING'].dropna(), hist=True, color='red')
plt.title(f'Distribution of RATING for Thriller')
plt.show()
```

The above histogram shows the ratings for the Action genre which has typically seems to receive higher ratings.While the thriller genre seemed popular too.

```python
top_genres = list(df['GENRE'].value_counts().head(6).index)
fig, ax = plt.subplots(2, 3, figsize=(10, 5))
for i, genre in enumerate(top_genres):
    filtered_df = df[df['GENRE'] == genre]
    row = i // 3
    col = i % 3
    ax[row, col].set_title(genre)
    ax[row, col].hist(filtered_df['RATING'], bins=10)
plt.tight_layout()
plt.show()
```
