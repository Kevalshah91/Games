```python
def initialize_board():
    return [[" " for _ in range(3)] for _ in range(3)]

def display_board(board):
    for row in board:
        print(" | ".join(row))
        print("-"*9)

def get_player_names():
    player1_name = input("Enter your name Player 1 (X): ")
    player2_name = input("Enter your name Player 2 (0): ")
    return player1_name, player2_name                

        
def make_move(player):
    box = int(input("Choose a box number (1-9): "))
    row = (box - 1) // 3
    col = (box - 1) % 3
    return row, col


def loc_move(board,row,col,player):
    board[row][col] = player
    
def check_win(board,player):
    for i in range(3):
        if all(board[i][j] == player for j in range(3)):
            return True
        if all(board[j][i] == player for j in range(3)):
            return True
        if all(board[i][i] == player for i in range(3)) or all(board[i][2-i] == player for i in range(3)):
            return True
    return False

def board_full(board):
    return all(board[i][j] != " " for i in range(3) for j in range(3))

def play_ttt():
    player1_name, player2_name = get_player_names()
    board = initialize_board()
    player_names = [player1_name,player2_name]
    players = ["X", "O"]
    player_now = 0

    while True:
        display_board(board)
        player = players[player_now]
        player_name = player_names[player_now]
        row, col = make_move(player_name)

        
        if board[row][col] == " ":
            loc_move(board,row,col,player)
            if check_win(board,player):
                display_board(board)
                print(f"{player_name} wins!")
                break
            elif board_full(board):
                display_board(board)
                print("It's a draw!")
                break
            else:
                player_now = (player_now + 1)%2
        else:
            print("Cell already occupied ")
 

if __name__ == "__main__":
    play_ttt()
```
